```python
# 第一题
import numpy as np
import math
def A(n):
    A = np.zeros((n, n))
    for i in range(n):
        for j in range(n):
            A[i, j] = 1 / (i + j + 1)
    return A


# Gauss 消去法
def Gauss(A, b):
    n = len(A)
    a = A.copy()
    b = b.copy()
    for i in range(n):
        for j in range(i + 1, n):
            factor = a[j, i] / a[i, i]
            a[j, i:] = a[j, i:] - factor * a[i, i:]
            b[j] = b[j] - factor * b[i]

    x = np.zeros(n)
    for i in range(n - 1, -1, -1):
        x[i] = (b[i] - np.dot(a[i, i + 1:], x[i + 1:])) / a[i, i]
    
    return x


# （2）
def D(n):
    D = np.zeros((n, n))
    for i in range(n):
        for j in range(n):
            if i == j:
                D[i, j] = np.sqrt( 1 / (i + 1) )
    return D

def D0_inv(n):
    D_inv = np.zeros((n, n))
    for i in range(n):
        for j in range(n):
            if i == j:
                D_inv[i, j] = np.sqrt(i + 1)
    return D_inv



n = 12
x_correct = np.array([1 for i in range(n)])
A = A(n)
b = A @ x_correct


# 1
x1 = Gauss(A, b)
result1 = np.linalg.norm(x1-x_correct)
#print(x1)
print(result1)


# 2
D = D(n)
D_inv = D0_inv(n)
B = D_inv @ A @ D_inv.T
d = D_inv @ b


y = Gauss(B, d)
x2 = D_inv.T @ y
result2 = np.linalg.norm(x2-x_correct)
#print(x2)
print(result2)
```

```python
#  复化求积
import math


# 复化梯形法
def trapezoidal(f, a, b, n):
    h = (b - a) / n
    s0 = 0.5 * (f(a) + f(b))
    s1 = 0
    for i in range(1, n):
        s1 += f(a + i * h)
    
    return h * (s0 + s1)


# 复化Simpson法
def simpson(f, a, b, n):
    h = (b - a) / n
    s0 = f(a) + f(b)
    s1 = 0
    s2 = 0
    for i in range(1, n):
        s1 += f(a + i * h)

    for i in range(0, n):
        s2 += f(a + i * h + h / 2)

    return h / 6 * (s0 + 2 * s1 + 4 * s2)



def f(x):
    return math.exp(x)*math.cos(x)

a = 0
b = math.pi
I = -12.070346316389634503

n = 32

Tn = trapezoidal(f, a, b, n)
Sn = simpson(f, a, b, n)
error1 = abs(I - Tn)
error2 = abs(I - Sn)

print("复化梯形法结果:", Tn)
print("复化梯形法误差:", error1)
print("复化Simpson法结果:", Sn)
print("复化Simpson法误差:", error2)
```

```python
# newton
import math

def Newton(f, df, x0, TOL, k=1):
    X = [x0]
    x = x0
    i = 0
    error = float('inf')
    while error > TOL:
        x = x - k * f(x) / df(x)
        X.append(x)
        i = i + 1
        error = abs(X[i]-X[i-1])
        print(f"迭代次数: {i}, 当前误差: {error}")

    return x,X


def f(x):
    return x**2 - 3*x - math.exp(x) + 2

def df(x):
    return 2*x - 3 - math.exp(x)


x0 = 1.0
TOL = 1e-8

x, X = Newton(f, df, x0, TOL)


print(X[-2:])
```
