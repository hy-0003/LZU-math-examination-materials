```python
# Gauss 消去法
import numpy as np

def A(n):
    A = np.zeros((n, n))
    for i in range(n):
        for j in range(n):
            A[i, j] = 1 / (i + j + 1)
    return A

n1 = 10
n2 = 100

x_correct1 = np.array([1 for i in range(n1)])
x_correct2 = np.array([1 for i in range(n2)])
A1 = A(n1)
A2 = A(n2)
b1 = A1 @ x_correct1
b2 = A2 @ x_correct2

def Gauss(A, b, x0):
    n = len(A)
    a = A
    for i in range(n):
        for j in range(i + 1, n):
            factor = a[j, i] / a[i, i]
            a[j, i:] = a[j, i:] - factor * a[i, i:]
            b[j] = b[j] - factor * b[i]

    x = np.zeros(n)
    for i in range(n - 1, -1, -1):
        x[i] = (b[i] - np.dot(a[i, i + 1:], x[i + 1:])) / a[i, i]
    
    return x
        

x1 = Gauss(A1, b1, x_correct1)
print(x1)
x2 = Gauss(A2, b2, x_correct2)
print(x2)
```
