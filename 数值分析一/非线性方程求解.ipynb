```python
#  二分法


def slove(a_0,b_0,f,epslion,delta,N):
    if f(a_0) == 0:
        print("a is the root")
        return a_0
    if f(b_0) == 0:
        print("b is the root")
        return b_0
    a = []
    b = []
    x = []
    a.append(a_0)
    b.append(b_0)
    x.append( (a[0] + b[0]) / 2 )
    for i in range(N):
        if abs(b[i] - x[i]) <= epslion or abs(f(x[i])) <= delta:
            return x , i ,a ,b
        else:
            if f(a[i]) * f(x[i]) > 0:
                a.append(x[i])
                b.append(b[i])
            else:
                a.append(a[i])
                b.append(x[i])
            x.append( (a[i+1] + b[i+1]) / 2 )
    print("No root found within the given iterations")


def print_table(result, f):
    if result is None or isinstance(result, (int, float)):
        return
    x, iterations, a, b = result
    print("迭代次数:", iterations)
    # 打印表头
    print(f"{'[a, b]':<15}{'x':<15}{'f(x)':<15}")
    print("-" * 45)
    for i in range(iterations + 1):
        # 打印每一行数据
        print(f"[{a[i]:.6f}, {b[i]:.6f}]{x[i]:<15.6f}{f(x[i]):<15.6f}")


def f(x):
    return x**3 + 4*x**2 - 10

a = 1
b = 2
epslion = 0.00001
delta = 0.00001
N = 1000

result = slove(a, b, f, epslion, delta, N)
print_table(result, f)
        
```

```python
# 简单迭代法
import cmath

def simple_iteration(f, x0, epslion, N):
    x = [x0]
    for i in range(N):
        try:
            next_x = f(x[i])
            x.append(next_x)
            if abs(next_x - x[i]) < epslion:  # 修改判断条件
                print(f"找到根：{next_x}，迭代次数：{i + 1}")
                return i, x
        except OverflowError:
            print(f"迭代到第 {i} 次时数值过大，无法继续计算。当前 x 值大致为 {x[i]}")
            return i, x
    print("No root found within the given iterations")
    return None

def print_table(result):
    if result is None or isinstance(result, (int, float)):
        return
    iterations, x = result
    print("迭代次数:", iterations)
    # 打印表头
    print(f"{'迭代':<10}{'x':<20}")
    print("-" * 30)
    for i in range(iterations + 1):
        try:
            print(f"{i:<10}{x[i]:<20.10f}")
        except OverflowError:
            print(f"{i:<10}{'数值过大':<20}")


def f1(x):
    return x - x**3 -4*x**2 + 10

def f2(x):
    return 0.5 * cmath.sqrt(10 - x**3)

def f3(x):
    return cmath.sqrt(10/x - 4*x)

def f4(x):
    return cmath.sqrt( 10/(4+x) )

def f5(x):
    return x - (x**3 + 4*x**2 - 10) / (3*x**2 + 8*x)



result1 = simple_iteration(f1, 1.5, 0.01, 10)
result2 = simple_iteration(f2, 1.5, 0.01, 10)
result3 = simple_iteration(f3, 1.5, 0.01, 10)
result4 = simple_iteration(f4, 1.5, 0.01, 10)
result5 = simple_iteration(f5, 1.5, 0.01, 10)


print("\n")
print_table(result1)
print_table(result2)
print_table(result3)
print_table(result4)
print_table(result5)
```

```python
# 牛顿迭代法
import cmath


def newton(f, df, x0, TOL, iterations, k=1):
    x = [x0]
    for i in range(iterations):
        x.append(x[i] - k * f(x[i]) / df(x[i]))
        if abs(f(x[i + 1])) < TOL:
            return x , i
    print("Newton's method did not converge after", iterations, "iterations.")


def Newton(f, df, x0, TOL, k=1):
    X = [x0]
    x = x0
    while abs(f(x)) > TOL:
        x = x - k * f(x) / df(x)
        X.append(x)

    return x,X


def f(x):
    return x**2 - 3*x - cmath.exp(x) + 2

def df(x):
    return 2*x - 3 - cmath.exp(x)


x0 = 1
TOL = 1e-8
iterations = 10000

x, i = newton(f, df, x0, TOL, iterations)
x_while, X_while = Newton(f, df, x0, TOL)
print(x[i])
print(x_while)


alpha = 7
def cf(x, alpha=7):
    return -x + alpha*x**2

def cdf(x):
    return 1.0


x0 = 0.01
TOL = 1e-8
x ,i = newton(cf, cdf, x0, TOL, iterations)
print(x[i])
```
