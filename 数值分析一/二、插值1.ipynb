```python
#lagrange插值
from math import *
import matplotlib.pyplot as plt



s=input("请输入x0,y0,x1,y1,x2......")
num=s.split(',')
l=len(num)

x=[]
y=[]
for i in range(l):
    if i % 2 == 0:
        x.append(float(num[i]))
    if i % 2 == 1:
        y.append(float(num[i]))


def lx(x,i):
    sum = 1
    for j in range(len(x)):
        if j == i:
            sum = sum
        else:
            sum = sum * (x[i]-x[j])
    return sum

def ls(ver,x,i):
    sum = 1
    for j in range(len(x)):
        if j == i:
            sum = sum
        else:
            sum = sum * (ver-x[j])
    return sum



def lagrange(ver,x,y):
    n=len(x)
    sum=0
    for i in range(n):
        sum = sum + y[i]*ls(ver,x,i)/lx(x,i)
    return sum   



ver=input("请输入想要的y对应的x")
ver = float(ver)
print(lagrange(ver,x,y))



x_min = min(x)-10
x_max = max(x)+10
x_range = x_max - x_min
x_values = [x_min + i * x_range / 100 for i in range(101)]  # 生成100个点

# 计算对应的y值
y_values = [lagrange(x_val, x, y) for x_val in x_values]

# 绘制图像
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, label="Lagrange Interpolation", color="blue")
plt.scatter(x, y, label="Given Data Points", color="red", zorder=5)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Lagrange Interpolation")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Neville插值
#导入包
from math import *
import matplotlib.pyplot as plt


#处理输入
s=input("请输入x0,y0,x1,y1,x2......")
num=s.split(',')
l=len(num)

x=[]
y=[]
for i in range(l):
    if i % 2 == 0:
        x.append(float(num[i]))
    if i % 2 == 1:
        y.append(float(num[i]))




#实现函数:以递推的方式
def neville(x0,x,y):
    n = len(x)
    def f(a,b):
        if a == b:
            return y[a]
        else:
            return ( (x[b]-x0) * f(a,b-1) + (x0-x[a]) * f(a+1,b) )/(x[b]-x[a])
        
    return f(0,n-1)




#预测值（ver就是你想预测的y对应的x）
ver=input("请输入想要的y对应的x")
ver = float(ver)
print(neville(ver,x,y))


#作图代码
x_min = min(x)-10    #x轴向左延长10个单位
x_max = max(x)+10    #x轴向右延长10个单位
x_range = x_max - x_min
x_values = [x_min + i * x_range / 100 for i in range(101)]  # 生成100个点
y_values = [neville(x_val, x, y) for x_val in x_values]    # 计算对应的y值

# 绘制图像
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, label="Neville Interpolation", color="blue")
plt.scatter(x, y, label="Given Data Points", color="red", zorder=5)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Neville Interpolation")
plt.legend()
plt.grid(True)
plt.show()
```
