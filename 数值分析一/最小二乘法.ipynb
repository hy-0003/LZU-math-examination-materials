```python
import numpy as np
import matplotlib.pyplot as plt


x = input("请输入x的值,用,分隔：").split(',')
y = input("请输入y的值,用,分隔：").split(',')
n = int(input("请输入n的值:"))

x = [float(i) for i in x]
y = [float(i) for i in y]
if len(x) != len(y):
    print("x和y的长度不一致,请重新输入。")
    exit()


def phi_i(x, i):
    return [x[j] ** i for j in range(len(x))]


def get_Y(y, x, n):
    return [np.dot(y, phi_i(x, i)) for i in range(n)]


def get_G(x, n):
    G = np.zeros((n, n))
    for i in range(n):
        for j in range(n):
            G[i][j] = np.dot(phi_i(x, i), phi_i(x, j))
    return G


Y = np.array(get_Y(y, x, n))
G = get_G(x, n)

try:
    c = np.linalg.solve(G, Y)
    print("求解得到的向量 c 为:", c)
except np.linalg.LinAlgError:
    print("系数矩阵 G 是奇异矩阵，无法求解线性方程组。")

p = np.poly1d(c[::-1])

# 绘制数据点和多项式函数
plt.scatter(x, y, label='Data Points', color='red')
x_vals = np.linspace(min(x)-5, max(x)+5, 400)
y_vals = p(x_vals)
plt.plot(x_vals, y_vals, label='Polynomial Function', color='blue')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Data Points and Polynomial Function')
plt.legend()
plt.grid(True)
plt.show()
```
