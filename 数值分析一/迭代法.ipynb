```python
#  追赶法
import numpy as np

def zg(A,b):
    n = len(A)
    e = []
    d = []
    f = []
    l = []
    for i in range(n-1):
        e.append(A[i][i])
        d.append(A[i+1][i])
        f.append(A[i][i+1])
    e.append(A[n-1][n-1])

    b = np.asarray(b, dtype=float).copy()

    for i in range(n-1):
        l.append(d[i]/(e[i]))
        e[i+1] = e[i+1] - l[i]*f[i]

    for i in range(1,n):
        b[i] = b[i] - l[i-1]*b[i-1]
    
    x = [0 for _ in range(n)]
    x[-1] = b[-1]/e[-1]
    for i in range(n-2,-1,-1):
        x[i] = (b[i]-f[i]*x[i+1])/e[i]

    return x



A = np.array([[2,1,0,0],[1,3,1,0],[0,1,1,1],[0,0,2,1]])
b = np.array([1,2,2,0])
x = zg(A,b)
print(x)

print(np.linalg.solve(A, [1.0,2,2,0])) 
```
